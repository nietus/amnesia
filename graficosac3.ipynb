# Associatividade

```python
import matplotlib.pyplot as plt
import numpy as np

associatividade = [1, 2, 4, 8]
tr_aleatorio = [189, 189, 189, 189]
tr_espacial = [189, 189, 189, 189]
tr_temporal = [69, 69, 69, 69]
miss_aleatorio = [56, 56, 56, 56]
miss_espacial = [56, 56, 56, 56]
miss_temporal = [14, 14, 14, 14]

plt.style.use('default')
plt.rcParams.update({
    'font.size': 10,
    'axes.titlesize': 11,
    'axes.labelsize': 10,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'grid.color': '#DDDDDD',
    'grid.linestyle': ':',
    'grid.alpha': 0.7,
    'figure.facecolor': 'white',
    'legend.framealpha': 1,
    'legend.fontsize': 9
})

colors = ['#4C72B0', '#55A868', '#C44E52']
width = 0.28
x = np.arange(len(associatividade))

# Gráfico 1: Tempo
fig1, ax1 = plt.subplots(figsize=(8, 4))
ax1.bar(x - width, tr_aleatorio, width, label='Trace Aleatório', color=colors[0])
ax1.bar(x, tr_espacial, width, label='Trace Espacial', color=colors[1])
ax1.bar(x + width, tr_temporal, width, label='Trace Temporal', color=colors[2])
ax1.set_ylabel('Tempo (ciclos)')
ax1.set_xlabel('Associatividade')
ax1.set_xticks(x)
ax1.set_xticklabels(associatividade)
ax1.grid(True, linestyle=':', alpha=0.7)
ax1.set_title('Aumento da Associatividade - Tempo')
ax1.set_ylim(0, max(tr_aleatorio) * 1.4)
ax1.legend(loc='upper right')
fig1.tight_layout()
fig1.savefig('tempo_acesso.png', dpi=300, bbox_inches='tight')
plt.show()

# Gráfico 2: Miss
fig2, ax2 = plt.subplots(figsize=(8, 4))
ax2.bar(x - width, miss_aleatorio, width, label='Trace Aleatório', color=colors[0])
ax2.bar(x, miss_espacial, width, label='Trace Espacial', color=colors[1])
ax2.bar(x + width, miss_temporal, width, label='Trace Temporal', color=colors[2])
ax2.set_ylabel('Taxa de Miss (%)')
ax2.set_xlabel('Associatividade')
ax2.set_xticks(x)
ax2.set_xticklabels(associatividade)
ax2.grid(True, linestyle=':', alpha=0.7)
ax2.set_title('Aumento da Associatividade - Taxa de Miss')
ax2.set_ylim(0, max(miss_aleatorio) * 1.4)
ax2.legend(loc='upper right')
fig2.tight_layout()
fig2.savefig('taxa_miss.png', dpi=300, bbox_inches='tight')
plt.show()
```
